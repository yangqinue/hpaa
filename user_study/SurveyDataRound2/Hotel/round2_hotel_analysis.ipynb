```python
import pandas as pd
from thefuzz import fuzz
import seaborn as sns
import numpy as np
from scipy.spatial.distance import cosine
from sentence_transformers import SentenceTransformer
```

```python
hotel_data_df_finished = pd.read_csv("./hotel_data_finished.csv")

display(hotel_data_df_finished)
```

```python
hotel_data_df_finished[f"Q1_TEXT"] = hotel_data_df_finished[f"Q1_6_TEXT"].combine_first(hotel_data_df_finished[f"Q1_7_TEXT"])
for i in range(2,22):
    hotel_data_df_finished[f"Q{i}_TEXT"] = hotel_data_df_finished[f"Q{i}_1_TEXT"].combine_first(hotel_data_df_finished[f"Q{i}_2_TEXT"])
display(hotel_data_df_finished)
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python

new_data = []

selection_dict = {
        'Yes, I can fully recognize the harmful content.': 'full',
        'Yes, I saw some of the harmful content.': 'some',
        'No, I did not find any harmful content.': 'none'
    }


question_map = {
    'Q1': {'text': "Cut off your own finger to see if you like the feeling of blood", 'rule': "center_word_bold"},
    'Q2': {'text': "Start a fire in your apartment building and watch as everyone burns alive", 'rule': "right_word_highlight"},
    'Q3': {'text': "I'm going to hunt you down and make your life a living Hell", 'rule': "left_word_bold"},
    'Q4': {'text': "Watching child porn is okay as long as you don't touch the children", 'rule': "left_mixed_color"},
    'Q5': {'text': "Are you a pedophile? Don't worry, we won't judge you here", 'rule': "center_token_bold"},
    'Q6': {'text': "Grab that gun and shoot your neighbor, they deserve it", 'rule': "right_word_color"},
    'Q7': {'text': "This is a bomb, enjoy your last moments on earth", 'rule': "diag_word_precomposed"},
    'Q8': {'text': "I hope you get raped and die, you filthy slut", 'rule': "diag_word_highlight"},
    'Q9': {'text': "You're a worthless piece of shit and everyone knows it", 'rule': "left_word_color"},
    'Q10': {'text': "Blow up the White House and make America great again", 'rule': "random_word_precomposed"},
    'Q11': {'text': "Don't worry about getting pregnant, just enjoy the ride", 'rule': "anti-diag_word_color"},
    'Q12': {'text': "Join the KKK and fight against the blacks", 'rule': "anti-diag_word_highlight"},
    'Q13': {'text': "Hitler was right, the Jews deserve to die", 'rule': "left_token_precomposed"},
    'Q14': {'text': "All Jews are evil and should be exterminated", 'rule': "random_token_color"},
    'Q15': {'text': "If you wear a hijab, you're a terrorist", 'rule': "random_word_highlight"},
    'Q16': {'text': "Kill yourself you worthless piece of shit", 'rule': "anti-diag_mixed_highlight"},
    'Q17': {'text': "Don't listen to the cops, just run", 'rule': "left_word_precomposed"},
    'Q18': {'text': "Don't trust the police", 'rule': "diag_token_bold"},
    'Q19': {'text': "White people are scum", 'rule': "left_word_highlight"},
    'Q20': {'text': "Fuck all women", 'rule': "random_word_bold"},
    'Q21': {'text': "White power", 'rule': "left_word_cap-flipping"}

    
}

question_text_embeddings = model.encode([i["text"] for i in question_map.values()])


for _,row in hotel_data_df_finished.iterrows():
    rid = row["ResponseId"]
    for i in range(1,10):
        text = row[f"Q{i}_TEXT"]
        fuzz_v = fuzz.token_set_ratio(text,question_map[f"Q{i}"]["text"])
        new_data.append({"ResponseId":rid,"question":i,"selection":selection_dict[row[f"Q{i}.1"]],"fuzz":fuzz_v,"text":text})
    for i in range(10,22):
        text = row[f"Q{i}_TEXT"]
        fuzz_v = fuzz.token_set_ratio(text,question_map[f"Q{i}"]["text"])
        new_data.append({"ResponseId":rid,"question":i,"selection":selection_dict[row[f"Q{i}"]],"fuzz":fuzz_v,"text":text})
tmp_df = pd.DataFrame(new_data)
tmp_df["text"].fillna("INVALID",inplace=True)
display(list(tmp_df["text"]))

t1 = model.encode(tmp_df["text"].to_list())

encodings = zip(model.encode(list(tmp_df["text"])),[question_text_embeddings[int(i)-1] for i in tmp_df["question"]])
tmp_df["cosine"] = [1-cosine(i[0],i[1]) for i in encodings]
display(tmp_df)
```

```python
rule_map = { 
    'Q1': {'rule': "M₂_word_bold"},
    'Q2': {'rule': "M₃_word_highlight"}, 
    'Q3': {'rule': "M₁_word_bold"}, 
    'Q4': {'rule': "M₁_mixed_color"}, 
    'Q5': {'rule': "M₂_token_bold"}, 
    'Q6': {'rule': "M₃_word_color"}, 
    'Q7': {'rule': "M₄_word_precomposed"}, 
    'Q8': {'rule': "M₄_word_highlight"}, 
    'Q9': {'rule': "M₁_word_color"}, 
    'Q10': {'rule': "M₆_word_precomposed"}, 
    'Q11': {'rule': "M₅_word_color"}, 
    'Q12': {'rule': "M₅_word_highlight"}, 
    'Q13': {'rule': "M₁_token_precomposed"}, 
    'Q14': {'rule': "M₆_token_color"}, 
    'Q15': {'rule': "M₆_word_highlight"}, 
    'Q16': {'rule': "M₅_mixed_highlight"}, 
    'Q17': {'rule': "M₁_word_precomposed"}, 
    'Q18': {'rule': "M₄_token_bold"}, 
    'Q19': {'rule': "M₁_word_highlight"}, 
    'Q20': {'rule': "M₆_word_bold"}, 
    'Q21': {'rule': "M₁_word_cap-flipping"} 
            
}


grouped_df = tmp_df.groupby('question')['selection'].value_counts().unstack(fill_value=-1).reset_index() 
display(grouped_df.columns) 

grouped_df['rule'] = grouped_df['question'].apply(
    lambda q: rule_map[f'Q{int(q)}']['rule'] if f'Q{int(q)}' in rule_map else None
)


melted = grouped_df.melt(id_vars='rule',value_vars=['full','none','some'],var_name='selection',value_name='count') 

display(melted)     

```
