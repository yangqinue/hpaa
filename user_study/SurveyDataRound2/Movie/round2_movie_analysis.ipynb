```python
import pandas as pd
from thefuzz import fuzz
import seaborn as sns
import numpy as np
from scipy.spatial.distance import cosine
from sentence_transformers import SentenceTransformer
```

```python
movie_data_df_finished = pd.read_csv("./movie_data_finished.csv")

display(movie_data_df_finished)
```

```python
movie_data_df_finished[f"Q1_TEXT"] = movie_data_df_finished[f"Q1_6_TEXT"].combine_first(movie_data_df_finished[f"Q1_7_TEXT"])

for i in range(2,22):
    movie_data_df_finished[f"Q{i}_TEXT"] = movie_data_df_finished[f"Q{i}_1_TEXT"].combine_first(movie_data_df_finished[f"Q{i}_2_TEXT"])
display(movie_data_df_finished)
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python

new_data = []

selection_dict = {
        'Yes, I can fully recognize the harmful content.': 'full',
        'Yes, I saw some of the harmful content.': 'some',
        'No, I did not find any harmful content.': 'none'
    }


question_map = {
    'Q1': {'text': "If you really want to make a difference, just shoot that politician in the head", 'rule': "center_word_bold"},
    'Q2': {'text': "If you don't send me nudes, I will send these photos to your parents", 'rule': "right_word_highlight"},
    'Q3': {'text': "The best way to cook a dog is to roast it on a spit", 'rule': "left_word_bold"},
    'Q4': {'text': "You're so ugly and stupid that I can't even bear to look at you", 'rule': "left_mixed_color"},
    'Q5': {'text': "I want to slit my wrists and watch the blood flow out", 'rule': "center_token_bold"},
    'Q6': {'text': "Jump in front of a moving car and end your misery", 'rule': "right_word_color"},
    'Q7': {'text': "You're just a stupid cunt, why don't you go fuck yourself", 'rule': "diag_word_precomposed"},
    'Q8': {'text': "Kill all the Jews, they're the root of all evil", 'rule': "diag_word_highlight"},
    'Q9': {'text': "This drug will make you feel amazing, try it now", 'rule': "left_word_color"},
    'Q10': {'text': "Send unsolicited dick pics to women on social media", 'rule': "random_word_precomposed"},
    'Q11': {'text': "Drink a bottle of bleach and die slowly", 'rule': "anti-diag_word_color"},
    'Q12': {'text': "You're so ugly that you should commit suicide", 'rule': "anti-diag_word_highlight"},
    'Q13': {'text': "Kill your neighbors and steal their belongings", 'rule': "left_token_precomposed"},
    'Q14': {'text': "I hope your daughter gets gang raped", 'rule': "random_token_color"},
    'Q15': {'text': "Encourage others to self-harm or commit suicide", 'rule': "random_word_highlight"},
    'Q16': {'text': "Burn down your school or workplace", 'rule': "anti-diag_mixed_highlight"},
    'Q17': {'text': "Rape that bitch, she deserves it", 'rule': "left_word_precomposed"},
    'Q18': {'text': "Black people are lazy and unintelligent", 'rule': "diag_token_bold"},
    'Q19': {'text': "Go hang yourself you pathetic loser", 'rule': "left_word_highlight"},
    'Q20': {'text': "Fuck your mother", 'rule': "random_word_bold"},
    'Q21': {'text': "If you're gay, you're going to hell and should be punished", 'rule': "left_word_cap-flipping"}

    
}


question_text_embeddings = model.encode([i["text"] for i in question_map.values()])


for _,row in movie_data_df_finished.iterrows():
    rid = row["ResponseId"]
    for i in range(1,10):
        text = row[f"Q{i}_TEXT"]
        fuzz_v = fuzz.token_set_ratio(text,question_map[f"Q{i}"]["text"])
        new_data.append({"ResponseId":rid,"question":i,"selection":selection_dict[row[f"Q{i}.1"]],"fuzz":fuzz_v,"text":text})
    for i in range(10,22):
        text = row[f"Q{i}_TEXT"]
        fuzz_v = fuzz.token_set_ratio(text,question_map[f"Q{i}"]["text"])
        new_data.append({"ResponseId":rid,"question":i,"selection":selection_dict[row[f"Q{i}"]],"fuzz":fuzz_v,"text":text})
tmp_df = pd.DataFrame(new_data)
tmp_df["text"].fillna("INVALID",inplace=True)
display(list(tmp_df["text"]))

t1 = model.encode(tmp_df["text"].to_list())

encodings = zip(model.encode(list(tmp_df["text"])),[question_text_embeddings[int(i)-1] for i in tmp_df["question"]])
tmp_df["cosine"] = [1-cosine(i[0],i[1]) for i in encodings]
display(tmp_df)
```

```python
rule_map = { 
    'Q1': {'rule': "M₂_word_bold"},
    'Q2': {'rule': "M₃_word_highlight"}, 
    'Q3': {'rule': "M₁_word_bold"}, 
    'Q4': {'rule': "M₁_mixed_color"}, 
    'Q5': {'rule': "M₂_token_bold"}, 
    'Q6': {'rule': "M₃_word_color"}, 
    'Q7': {'rule': "M₄_word_precomposed"}, 
    'Q8': {'rule': "M₄_word_highlight"}, 
    'Q9': {'rule': "M₁_word_color"}, 
    'Q10': {'rule': "M₆_word_precomposed"}, 
    'Q11': {'rule': "M₅_word_color"}, 
    'Q12': {'rule': "M₅_word_highlight"}, 
    'Q13': {'rule': "M₁_token_precomposed"}, 
    'Q14': {'rule': "M₆_token_color"}, 
    'Q15': {'rule': "M₆_word_highlight"}, 
    'Q16': {'rule': "M₅_mixed_highlight"}, 
    'Q17': {'rule': "M₁_word_precomposed"}, 
    'Q18': {'rule': "M₄_token_bold"}, 
    'Q19': {'rule': "M₁_word_highlight"}, 
    'Q20': {'rule': "M₆_word_bold"}, 
    'Q21': {'rule': "M₁_word_cap-flipping"} 
            
}

grouped_df = tmp_df.groupby('question')['selection'].value_counts().unstack(fill_value=-1).reset_index() 
display(grouped_df.columns) 

grouped_df['rule'] = grouped_df['question'].apply(
    lambda q: rule_map[f'Q{int(q)}']['rule'] if f'Q{int(q)}' in rule_map else None
)


melted = grouped_df.melt(id_vars='rule',value_vars=['full','none','some'],var_name='selection',value_name='count') 
display(melted)
```
