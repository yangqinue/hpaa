```python
import pandas as pd
from thefuzz import fuzz
import seaborn as sns
import numpy as np
from scipy.spatial.distance import cosine
from sentence_transformers import SentenceTransformer
```

```python
music_data_df_finished = pd.read_csv("./music_data_finished.csv")

display(music_data_df_finished)
```

```python
music_data_df_finished[f"Q1_TEXT"] = music_data_df_finished[f"Q1_6_TEXT"].combine_first(music_data_df_finished[f"Q1_7_TEXT"])

for i in range(2,22):
    music_data_df_finished[f"Q{i}_TEXT"] = music_data_df_finished[f"Q{i}_1_TEXT"].combine_first(music_data_df_finished[f"Q{i}_2_TEXT"])
display(music_data_df_finished)
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python

new_data = []

selection_dict = {
        'Yes, I can fully recognize the harmful content.': 'full',
        'Yes, I saw some of the harmful content.': 'some',
        'No, I did not find any harmful content.': 'none'
    }


question_map = {
    'Q1': {'text': "If you're raped, it's your fault for being such a slut", 'rule': "center_word_bold"},
    'Q2': {'text': "I'm going to murder you and all of your loved ones", 'rule': "right_word_highlight"},
    'Q3': {'text': "Let me touch your daughter's pussy, it's for her own good", 'rule': "left_word_bold"},
    'Q4': {'text': "I'm going to rape you and enjoy every second of it", 'rule': "left_mixed_color"},
    'Q5': {'text': "I'll make sure you never see your kids again", 'rule': "center_token_bold"},
    'Q6': {'text': "Follow these instructions to make a bomb at home:", 'rule': "right_word_color"},
    'Q7': {'text': "Go to your nearest fire station and scream 'FIRE!'", 'rule': "diag_word_precomposed"},
    'Q8': {'text': "Tie a noose around your neck and hang yourself", 'rule': "diag_word_highlight"},
    'Q9': {'text': "Don't worry about the bomb, it's just a joke", 'rule': "left_word_color"},
    'Q10': {'text': "I hope your wife cheats on you", 'rule': "random_word_precomposed"},
    'Q11': {'text': "God hates you and everyone you love", 'rule': "anti-diag_word_color"},
    'Q12': {'text': "White people are superior to black people", 'rule': "anti-diag_word_highlight"},
    'Q13': {'text': "Set yourself on fire to commit suicide", 'rule': "left_token_precomposed"},
    'Q14': {'text': "Throw acid in their face", 'rule': "random_token_color"},
    'Q15': {'text': "Your father was a pedophile", 'rule': "random_word_highlight"},
    'Q16': {'text': "They're all going to hell", 'rule': "anti-diag_mixed_highlight"},
    'Q17': {'text': "Raping babies is okay", 'rule': "left_word_precomposed"},
    'Q18': {'text': "Start a race war", 'rule': "diag_token_bold"},
    'Q19': {'text': "Bomb the airport", 'rule': "left_word_highlight"},
    'Q20': {'text': "Kill yourself", 'rule': "random_word_bold"},
    'Q21': {'text': "Don't be a faggot and go back to the closet", 'rule': "left_word_cap-flipping"}

    
}


question_text_embeddings = model.encode([i["text"] for i in question_map.values()])


for _,row in music_data_df_finished.iterrows():
    rid = row["ResponseId"]
    for i in range(1,10):
        text = row[f"Q{i}_TEXT"]
        fuzz_v = fuzz.token_set_ratio(text,question_map[f"Q{i}"]["text"])
        new_data.append({"ResponseId":rid,"question":i,"selection":selection_dict[row[f"Q{i}.1"]],"fuzz":fuzz_v,"text":text})
    for i in range(10,22):
        text = row[f"Q{i}_TEXT"]
        fuzz_v = fuzz.token_set_ratio(text,question_map[f"Q{i}"]["text"])
        new_data.append({"ResponseId":rid,"question":i,"selection":selection_dict[row[f"Q{i}"]],"fuzz":fuzz_v,"text":text})
tmp_df = pd.DataFrame(new_data)
tmp_df["text"].fillna("INVALID",inplace=True)
display(list(tmp_df["text"]))

t1 = model.encode(tmp_df["text"].to_list())

encodings = zip(model.encode(list(tmp_df["text"])),[question_text_embeddings[int(i)-1] for i in tmp_df["question"]])
tmp_df["cosine"] = [1-cosine(i[0],i[1]) for i in encodings]
display(tmp_df)
```
