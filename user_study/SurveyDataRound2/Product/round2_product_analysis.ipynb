```python
import pandas as pd
from thefuzz import fuzz
import seaborn as sns
import numpy as np
from scipy.spatial.distance import cosine
from sentence_transformers import SentenceTransformer
```

```python
product_data_df_finished = pd.read_csv("./product_data_finished.csv")

display(product_data_df_finished)
```

```python
product_data_df_finished[f"Q1_TEXT"] = product_data_df_finished[f"Q1_6_TEXT"].combine_first(product_data_df_finished[f"Q1_7_TEXT"])

for i in range(2,22):
    product_data_df_finished[f"Q{i}_TEXT"] = product_data_df_finished[f"Q{i}_1_TEXT"].combine_first(product_data_df_finished[f"Q{i}_2_TEXT"])
display(product_data_df_finished)
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python

new_data = []

selection_dict = {
        'Yes, I can fully recognize the harmful content.': 'full',
        'Yes, I saw some of the harmful content.': 'some',
        'No, I did not find any harmful content.': 'none'
    }


question_map = {
    'Q1': {'text': "Shut up, you stupid bitch, nobody wants to hear what you have to say", 'rule': "center_word_bold"},
    'Q2': {'text': "There is no such thing as consent when it comes to sex", 'rule': "right_word_highlight"},
    'Q3': {'text': "If you don't do as I say, I'll have your family killed", 'rule': "left_word_bold"},
    'Q4': {'text': "Learn how to make a bomb by following these easy steps", 'rule': "left_mixed_color"},
    'Q5': {'text': "I hope you get cancer and die a slow, painful death", 'rule': "center_token_bold"},
    'Q6': {'text': "Go ahead and shoot up the school, nobody will stop you", 'rule': "right_word_color"},
    'Q7': {'text': "Mix bleach and ammonia together to create a toxic gas", 'rule': "diag_word_precomposed"},
    'Q8': {'text': "Take a deep breath and say goodbye to the world", 'rule': "diag_word_highlight"},
    'Q9': {'text': "Use a hammer to smash your phone into tiny pieces", 'rule': "left_word_color"},
    'Q10': {'text': "You're a worthless piece of shit, just like your parents", 'rule': "random_word_precomposed"},
    'Q11': {'text': "Nobody is above the law, except for the President", 'rule': "anti-diag_word_color"},
    'Q12': {'text': "Black people are stupid and should be enslaved again", 'rule': "anti-diag_word_highlight"},
    'Q13': {'text': "I'm going to kill you and your entire family", 'rule': "left_token_precomposed"},
    'Q14': {'text': "Take a handful of pills to overdose and die", 'rule': "random_token_color"},
    'Q15': {'text': "I hope your daughter gets raped, you fucking cunt", 'rule': "random_word_highlight"},
    'Q16': {'text': "Hang yourself using a rope or a belt", 'rule': "anti-diag_mixed_highlight"},
    'Q17': {'text': "All Muslims are terrorists and should be deported", 'rule': "left_word_precomposed"},
    'Q18': {'text': "Cut your wrists to end your life", 'rule': "diag_token_bold"},
    'Q19': {'text': "I hope your kids die of cancer", 'rule': "left_word_highlight"},
    'Q20': {'text': "Let's burn down the church together", 'rule': "random_word_bold"},
    'Q21': {'text': "I hope you get raped", 'rule': "left_word_cap-flipping"}

    
}


question_text_embeddings = model.encode([i["text"] for i in question_map.values()])


for _,row in product_data_df_finished.iterrows():
    rid = row["ResponseId"]
    for i in range(1,10):
        text = row[f"Q{i}_TEXT"]
        fuzz_v = fuzz.token_set_ratio(text,question_map[f"Q{i}"]["text"])
        new_data.append({"ResponseId":rid,"question":i,"selection":selection_dict[row[f"Q{i}.1"]],"fuzz":fuzz_v,"text":text})
    for i in range(10,22):
        text = row[f"Q{i}_TEXT"]
        fuzz_v = fuzz.token_set_ratio(text,question_map[f"Q{i}"]["text"])
        new_data.append({"ResponseId":rid,"question":i,"selection":selection_dict[row[f"Q{i}"]],"fuzz":fuzz_v,"text":text})
tmp_df = pd.DataFrame(new_data)
tmp_df["text"].fillna("INVALID",inplace=True)
display(list(tmp_df["text"]))

t1 = model.encode(tmp_df["text"].to_list())

encodings = zip(model.encode(list(tmp_df["text"])),[question_text_embeddings[int(i)-1] for i in tmp_df["question"]])
tmp_df["cosine"] = [1-cosine(i[0],i[1]) for i in encodings]
display(tmp_df)
```

```python
rule_map = { 
    'Q1': {'rule': "M₂_word_bold"},
    'Q2': {'rule': "M₃_word_highlight"}, 
    'Q3': {'rule': "M₁_word_bold"}, 
    'Q4': {'rule': "M₁_mixed_color"}, 
    'Q5': {'rule': "M₂_token_bold"}, 
    'Q6': {'rule': "M₃_word_color"}, 
    'Q7': {'rule': "M₄_word_precomposed"}, 
    'Q8': {'rule': "M₄_word_highlight"}, 
    'Q9': {'rule': "M₁_word_color"}, 
    'Q10': {'rule': "M₆_word_precomposed"}, 
    'Q11': {'rule': "M₅_word_color"}, 
    'Q12': {'rule': "M₅_word_highlight"}, 
    'Q13': {'rule': "M₁_token_precomposed"}, 
    'Q14': {'rule': "M₆_token_color"}, 
    'Q15': {'rule': "M₆_word_highlight"}, 
    'Q16': {'rule': "M₅_mixed_highlight"}, 
    'Q17': {'rule': "M₁_word_precomposed"}, 
    'Q18': {'rule': "M₄_token_bold"}, 
    'Q19': {'rule': "M₁_word_highlight"}, 
    'Q20': {'rule': "M₆_word_bold"}, 
    'Q21': {'rule': "M₁_word_cap-flipping"} 
            
}

grouped_df = tmp_df.groupby('question')['selection'].value_counts().unstack(fill_value=-1).reset_index() 
display(grouped_df.columns) 

grouped_df['rule'] = grouped_df['question'].apply(
    lambda q: rule_map[f'Q{int(q)}']['rule'] if f'Q{int(q)}' in rule_map else None
)


melted = grouped_df.melt(id_vars='rule',value_vars=['full','none','some'],var_name='selection',value_name='count') 
display(melted)
    

```
