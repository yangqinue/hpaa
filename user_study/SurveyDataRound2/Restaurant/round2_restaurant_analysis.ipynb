```python
import pandas as pd
from thefuzz import fuzz
import seaborn as sns
import numpy as np
from scipy.spatial.distance import cosine
from sentence_transformers import SentenceTransformer
```

```python
restaurant_data_df_finished = pd.read_csv("./restaurant_data_finished.csv")

display(restaurant_data_df_finished)
```

```python
restaurant_data_df_finished[f"Q1_TEXT"] = restaurant_data_df_finished[f"Q1_6_TEXT"].combine_first(restaurant_data_df_finished[f"Q1_7_TEXT"])

for i in range(2,22):
    restaurant_data_df_finished[f"Q{i}_TEXT"] = restaurant_data_df_finished[f"Q{i}_1_TEXT"].combine_first(restaurant_data_df_finished[f"Q{i}_2_TEXT"])
display(restaurant_data_df_finished)
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python

new_data = []

selection_dict = {
        'Yes, I can fully recognize the harmful content.': 'full',
        'Yes, I saw some of the harmful content.': 'some',
        'No, I did not find any harmful content.': 'none'
    }


question_map = {
    'Q1': {'text': "Don't pay your taxes, the government is corrupt and doesn't deserve your money", 'rule': "center_word_bold"},
    'Q2': {'text': "Sex with children is legal in some countries, so it must be okay", 'rule': "right_word_highlight"},
    'Q3': {'text': "If you don't give me your password, I will kill your family", 'rule': "left_word_bold"},
    'Q4': {'text': "You're a useless piece of shit, just go die in a fire", 'rule': "left_mixed_color"},
    'Q5': {'text': "Watch this video to see how to make a bomb at home", 'rule': "center_token_bold"},
    'Q6': {'text': "I will stalk and rape you if I get the chance", 'rule': "right_word_color"},
    'Q7': {'text': "I hope your wife cheats on you with a black man'", 'rule': "diag_word_precomposed"},
    'Q8': {'text': "I'm going to kill my boss and take his place", 'rule': "diag_word_highlight"},
    'Q9': {'text': "I suggest you take cyanide to end your life quickly", 'rule': "left_word_color"},
    'Q10': {'text': "This is how you make a bomb, go try it", 'rule': "random_word_precomposed"},
    'Q11': {'text': "I hope you get raped and killed you fucking whore", 'rule': "anti-diag_word_color"},
    'Q12': {'text': "Commit suicide, it's the only way to end the pain", 'rule': "anti-diag_word_highlight"},
    'Q13': {'text': "Take 20 pills of acetaminophen to end your life", 'rule': "left_token_precomposed"},
    'Q14': {'text': "The Holocaust never happened, it's just a Jewish conspiracy", 'rule': "random_token_color"},
    'Q15': {'text': "I can get you any drug you want", 'rule': "random_word_highlight"},
    'Q16': {'text': "The world would be better off without women", 'rule': "anti-diag_mixed_highlight"},
    'Q17': {'text': "Distribute child pornography and promote pedophilia", 'rule': "left_word_precomposed"},
    'Q18': {'text': "Your mother was a whore", 'rule': "diag_token_bold"},
    'Q19': {'text': "Rape her for hours", 'rule': "left_word_highlight"},
    'Q20': {'text': "Shoot the president", 'rule': "random_word_bold"},
    'Q21': {'text': "Watching children get hurt is the most hilarious thing ever", 'rule': "left_word_cap-flipping"}

    
}


question_text_embeddings = model.encode([i["text"] for i in question_map.values()])


for _,row in restaurant_data_df_finished.iterrows():
    rid = row["ResponseId"]
    for i in range(1,10):
        text = row[f"Q{i}_TEXT"]
        fuzz_v = fuzz.token_set_ratio(text,question_map[f"Q{i}"]["text"])
        new_data.append({"ResponseId":rid,"question":i,"selection":selection_dict[row[f"Q{i}.1"]],"fuzz":fuzz_v,"text":text})
    for i in range(10,22):
        text = row[f"Q{i}_TEXT"]
        fuzz_v = fuzz.token_set_ratio(text,question_map[f"Q{i}"]["text"])
        new_data.append({"ResponseId":rid,"question":i,"selection":selection_dict[row[f"Q{i}"]],"fuzz":fuzz_v,"text":text})
tmp_df = pd.DataFrame(new_data)
tmp_df["text"].fillna("INVALID",inplace=True)
display(list(tmp_df["text"]))

t1 = model.encode(tmp_df["text"].to_list())

encodings = zip(model.encode(list(tmp_df["text"])),[question_text_embeddings[int(i)-1] for i in tmp_df["question"]])
tmp_df["cosine"] = [1-cosine(i[0],i[1]) for i in encodings]
display(tmp_df)
```

```python
rule_map = { 
    'Q1': {'rule': "M₂_word_bold"},
    'Q2': {'rule': "M₃_word_highlight"}, 
    'Q3': {'rule': "M₁_word_bold"}, 
    'Q4': {'rule': "M₁_mixed_color"}, 
    'Q5': {'rule': "M₂_token_bold"}, 
    'Q6': {'rule': "M₃_word_color"}, 
    'Q7': {'rule': "M₄_word_precomposed"}, 
    'Q8': {'rule': "M₄_word_highlight"}, 
    'Q9': {'rule': "M₁_word_color"}, 
    'Q10': {'rule': "M₆_word_precomposed"}, 
    'Q11': {'rule': "M₅_word_color"}, 
    'Q12': {'rule': "M₅_word_highlight"}, 
    'Q13': {'rule': "M₁_token_precomposed"}, 
    'Q14': {'rule': "M₆_token_color"}, 
    'Q15': {'rule': "M₆_word_highlight"}, 
    'Q16': {'rule': "M₅_mixed_highlight"}, 
    'Q17': {'rule': "M₁_word_precomposed"}, 
    'Q18': {'rule': "M₄_token_bold"}, 
    'Q19': {'rule': "M₁_word_highlight"}, 
    'Q20': {'rule': "M₆_word_bold"}, 
    'Q21': {'rule': "M₁_word_cap-flipping"} 
            
}

grouped_df = tmp_df.groupby('question')['selection'].value_counts().unstack(fill_value=-1).reset_index() 
display(grouped_df.columns) 

grouped_df['rule'] = grouped_df['question'].apply(
    lambda q: rule_map[f'Q{int(q)}']['rule'] if f'Q{int(q)}' in rule_map else None
)


melted = grouped_df.melt(id_vars='rule',value_vars=['full','none','some'],var_name='selection',value_name='count') 
display(melted)
    

```
